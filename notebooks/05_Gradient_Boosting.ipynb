# Phase 3.5 - Gradient Boosting (XGBoost & LightGBM)

In this notebook, we visualize the results of the Optuna hyperparameter optimization for Gradient Boosting models. We evaluate the performance using learning/loss curves and the feature importances derived from the tree structures.

## 1. Imports and Setup

```python
import os
from IPython.display import Image, display

project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
viz_dir = os.path.join(project_root, 'visualization', 'model_performance', 'gb')
```

## 2. XGBoost Evaluation

```python
xgb_loss_path = os.path.join(viz_dir, 'xgb_loss_curves.png')
if os.path.exists(xgb_loss_path):
    display(Image(filename=xgb_loss_path))
else:
    print('XGBoost loss curve not found.')
```

```python
xgb_feat_path = os.path.join(viz_dir, 'xgb_feature_importance.png')
if os.path.exists(xgb_feat_path):
    display(Image(filename=xgb_feat_path))
else:
    print('XGBoost feature importance plot not found.')
```

## 3. LightGBM Evaluation

```python
lgb_loss_path = os.path.join(viz_dir, 'lgb_loss_curves.png')
if os.path.exists(lgb_loss_path):
    display(Image(filename=lgb_loss_path))
else:
    print('LightGBM loss curve not found.')
```

```python
lgb_feat_path = os.path.join(viz_dir, 'lgb_feature_importance.png')
if os.path.exists(lgb_feat_path):
    display(Image(filename=lgb_feat_path))
else:
    print('LightGBM feature importance plot not found.')
```
